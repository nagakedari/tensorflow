```python
import tensorflow as tf
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
import pathlib
import PIL

print(tf.__version__)
```

```python
#Download the pre existing dataset from google
dataset_url = 'https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz'
data_dir = keras.utils.get_file('flower_photos', origin=dataset_url, untar=True)
data_dir = pathlib.Path(data_dir)
```

```python
# get the image count
image_count = len(list(data_dir.glob('*/*.jpg')))
image_count
```

```python
roses = list(data_dir.glob('roses/*'))
PIL.Image.open(str(roses[1]))
```

```python
tulips = list(data_dir.glob('tulips/*'))
PIL.Image.open(str(tulips[1]))
```

```python
data_dir
```

```python
batch_size = 32
img_heigh = 180
img_width = 180

#split the data set in to 2. One for training the model and another one for validating the model

train_ds = tf.keras.preprocessing.image_dataset_from_directory(
        data_dir,
        validation_split=0.2,
        subset="training",
        seed=123,
        image_size=(img_heigh, img_width),
        batch_size=batch_size )
train_ds
```

```python
# dataset for validating the model
val_ds = keras.preprocessing.image_dataset_from_directory(
        data_dir,
        validation_split=0.2,
        subset='validation',
        seed=123,
        image_size=(img_heigh, img_width),
        batch_size=batch_size)
```

```python
#different class names in the dataset
class_names = train_ds.class_names
class_names
```

```python

#get the first 9 images from the training data set
plt.figure(figsize=(10,10))

for images, lables in train_ds.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i+1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[lables[i]])
        plt.axis("off")
```

```python
# The RGB channel values are in the [0, 255] range. This is not ideal for a neural network; 
#in general you should seek to make your input values small. Here, we will standardize values to be in the [0, 1] 
#by using a Rescaling layer.
normalization_layer = layers.experimental.preprocessing.Rescaling(1./255, input_shape=(img_heigh, img_width, 3))
```

```python
for image_batch, labels_batch in train_ds:
  print(image_batch.shape)
  print(labels_batch.shape)
  break

  
```

```python
#There are two ways to use this layer. You can apply it to the dataset by calling map:

#Or, you can include the layer inside your model definition to simplify deployment. We will use the second approach here.
normalized_ds = train_ds.map(lambda x, y: (normalization_layer(x),y))
image_batch, labels_batch = next(iter(normalized_ds))
first_image = image_batch[0]
# Notice the pixels values are now in `[0,1]`.
print(np.min(first_image), np.max(first_image))
```

```python
# caching the data set for performance.
# prefetch() overlaps data preprocessing and model execution while training.
AUTOTUNE = tf.data.experimental.AUTOTUNE
train_ds = train_ds.cache().prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

```python
# Create a model
model = Sequential([normalization_layer, 
                    layers.Conv2D(16,3,activation='relu', padding = 'same'),
                    layers.MaxPooling2D(),
                    layers.Conv2D(32,3,activation='relu', padding = 'same'),
                    layers.MaxPooling2D(),
                    layers.Conv2D(64,3,activation='relu', padding = 'same'),
                    layers.MaxPooling2D(),
                    layers.Flatten(),
                    layers.Dense(128, activation='relu'),
                    layers.Dense(len(class_names))
                   ])
```

```python
# Compile a Model
model.compile(
    optimizer='adam',
    loss=tf.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy'])
```

```python
# model summary
model.summary()
```

```python
# Train the model
epochs = 10
history = model.fit(
        train_ds,
        validation_data=val_ds,
        epochs=epochs
)
```

```python
#Visualize training results

train_accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']

train_loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(epochs)

plt.figure(figsize=(8,8))
plt.subplot(1,2,1) # subplot arguments nrows, ncolumns, index. Here 1 row, 2 column, first graph
plt.plot(epochs_range, train_accuracy, label='Training Accuracy')
plt.plot(epochs_range, val_accuracy, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1,2,2) # subplot arguments nrows, ncolumns, index. Here 1 row, 2 column, 2nd graph
plt.plot(epochs_range, train_loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')

plt.show()
```

As you can see from the plots, training accuracy and validation accuracy are off by large margin and the model has achieved only around 60% accuracy on the validation set.<br><br>

Let's look at what went wrong and try to increase the overall performance of the model.

<h2 id="overfitting" data-text="Overfitting">Overfitting</h2>

<p>In the plots above, the training accuracy is increasing linearly over time, whereas validation accuracy stalls around 60% in the training process. Also, the difference in accuracy between training and validation accuracy is noticeable—a sign of <a href="https://www.tensorflow.org/tutorials/keras/overfit_and_underfit">overfitting</a>.</p>

<p>When there are a <u><b>small number of training examples</b></u>, the model sometimes learns from <u><b>noises or unwanted details</b></u> from training examples—to an extent that it negatively impacts the performance of the model on new examples. This phenomenon is known as overfitting. It means that the model will have a difficult time generalizing on a new dataset.</p>

<p>There are multiple ways to fight overfitting in the training process. In this tutorial, you&#39;ll use <em>data augmentation</em> and add <em>Dropout</em> to your model.</p>

```python
# Data augmentation takes the approach of generating additional training data from your existing examples 
# by augmenting them using random transformations that yield believable-looking images
data_augmentation = Sequential([
    layers.experimental.preprocessing.RandomFlip("horizontal",
                                                input_shape=(img_heigh, img_width, 3)),
    layers.experimental.preprocessing.RandomRotation(0.1),
    layers.experimental.preprocessing.RandomZoom(0.1)
])
```

```python
#Let's visualize what a few augmented examples look like by applying data augmentation to the same image several times
plt.figure(figsize=(10,10))
for images, lables in train_ds.take(1):
    for i in range(9):
#         plt.imshow(images[0].numpy().astype("uint8"))
        augmented_image = data_augmentation(images)
        ax = plt.subplot(3, 3, i+1)
        plt.imshow(augmented_image[0].numpy().astype("uint8"))
        plt.axis("off")
```

```python
# model with data augmentation
model = Sequential([
  data_augmentation,
  layers.experimental.preprocessing.Rescaling(1./255),
  layers.Conv2D(16, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(32, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(64, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Dropout(0.2),
  layers.Flatten(),
  layers.Dense(128, activation='relu'),
  layers.Dense(len(class_names))
])
```

```python
# compile the augmented model
model.compile(optimizer='adam',
             loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
             metrics=['accuracy'])
```

```python
# Augmented model summary
model.summary()
```

```python
epochs = 15
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=epochs
)
```
